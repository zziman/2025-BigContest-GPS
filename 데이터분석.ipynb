```python
df = franchise_data.copy()

# 매출 상위 그룹: Top 20~25%
top_group = df[df["매출금액_구간"].isin(["1_10%이하", "2_10-25%"])]

# 매출 하위 그룹: Bottom 10%
bottom_group = df[df["매출금액_구간"] == "6_90%초과(하위 10% 이하)"]

print("전체 데이터 수:", len(df))
print("상위 그룹 수 (Top 20~25%):", len(top_group))
print("하위 그룹 수 (Bottom 10%):", len(bottom_group))
```

```python
metrics = [
    "단골손님_비중",
    "신규손님_비중",
    "배달매출_비중",
    "업종매출지수_백분위",
    "업종건수지수_백분위"
]

result = pd.DataFrame({
    "metric": metrics,
    "top_mean": [top_group[m].mean() for m in metrics],
    "top_std":  [top_group[m].std()  for m in metrics],
    "bottom_mean": [bottom_group[m].mean() for m in metrics],
    "bottom_std":  [bottom_group[m].std()  for m in metrics],
})

result
```

```python
thresholds = {
    m: {
        "upper": result[result["metric"] == m]["top_mean"].values[0] 
                  + result[result["metric"] == m]["top_std"].values[0],
        "lower": result[result["metric"] == m]["top_mean"].values[0] 
                  - result[result["metric"] == m]["top_std"].values[0],
    }
    for m in metrics
}
thresholds
```
